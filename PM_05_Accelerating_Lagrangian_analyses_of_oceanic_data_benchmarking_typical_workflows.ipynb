## Purpose

For data, *Lagrangian* refers to oceanic and atmosphere information acquired by observing platforms drifting with the flow they are embedded within, but also more broadly refers to the data originating from uncrewed platforms, vehicles, and animals that gather data along their unrestricted but complicated paths. Because such paths traverse both spatial and temporal dimensions, Lagrangian data often convolve spatial and temporal information that cannot always and readily be organized, cataloged, and stored in common data structures and file formats with the help of common libraries and standards. As such, for both data generators and data users, Lagrangian data present challenges that the [CloudDrift](https://github.com/Cloud-Drift) project aims to overcome.

This notebook consists of systematic comparisons and evaluations of workflows for Lagrangian data, using as a basis the velocity and sea surface temperature dataset emanating from the drifting buoys of the [Global Drifter Program](https://www.aoml.noaa.gov/phod/gdp/). Specifically, we consider the interplay between diverse storage file formats ([NetCDF](https://www.unidata.ucar.edu/software/netcdf/), [Parquet](https://github.com/apache/parquet-format)) and the data structure associated with common existing libraries in Python ([xarray](https://docs.xarray.dev/en/stable/), [pandas](https://pandas.pydata.org), and [awkward](https://awkward-array.org/quickstart.html)) in order to test their adequacies for performing three common Lagrangian tasks: (i) binning of a variable on an Eulerian grid (e.g. mean temperature map); (ii) extracting data within given geographical and/or temporal windows; and (iii) analyses per trajectory (e.g. single statistics, Fast Fourier Transforms). Because the CloudDrift project aims at accelerating the use of Lagrangian data for atmospheric, oceanic, and climate sciences, we hope that this notebook will incite the community of users to eventually test it on their own platform and thus provide us with feedback on its ease of use and the intuitiveness of the proposed methods and guide future developments.

# Data Overview

The first goal of the notebook is to present the current format of the [Global Drifter Program (GDP)](https://www.aoml.noaa.gov/phod/gdp/) dataset, and its transformation into a single archive where each variable is stored in an ragged fashion.

The GDP produces two interpolated datasets of drifter position, velocity and sea surface temperature from more than 20,000 drifters that have been released since 1979. One dataset is at 6-hour resolution ([Hansen and Poulain 1996](http://dx.doi.org/10.1175/1520-0426(1996)013<0900:QCAIOW>2.0.CO;2)) and the other one is at hourly resolution ([Elipot et al. 2016](http://dx.doi.org/10.1002/2016JC011716)). The files, one per drifter identified by its unique identification number (ID), are updated on a quarterly basis and are available via the FTP website of GDP Data Assembly Center.

Our Notebook used a subset of the drifters of the GDP. Set at `subset_nb_drifters = 500`, the number of trajectories is large enough to create an interesting dataset while speeding up the fetching and processing of the data. It is possible to scale this value. The netCDF files are downloaded from the official `ftp` server.

```python
subset_nb_drifters = 500  # you can scale up/down this number (maximum value of 17324)
```

```python
%%time
# load the complete list of drifter IDs
list_id = pd.read_csv('data/gdp_ids.csv')

# "randomly" select a subset of id
np.random.seed(42)  # reproducible results
subset_id = np.random.choice(list_id.values.flatten(), subset_nb_drifters, replace=False)

# output folder and official GDP ftp
folder =  'data/raw/'
input_url = 'ftp://ftp.aoml.noaa.gov/phod/pub/lumpkin/hourly/v2.00/netcdf/'

def fetch_netcdf(url, file):
    '''
    Download and save file from the given url (if not present)
    '''
    if not isfile(file):
        req = urllib.request.urlretrieve(url, file)

# ~2min for 500 drifters
with concurrent.futures.ThreadPoolExecutor() as exector:
    # create list of urls and paths
    urls = []
    files = []
    for i in subset_id:
        file = f'drifter_{i}.nc'
        urls.append(join(input_url, file))
        files.append(join(folder, file))
    
    # parallel retrieving of individial nc files 
    exector.map(fetch_netcdf, urls, files)
```

## Individual netCDFs

The Global drifter program organized file by the drifter unique number, its ID. A few times a year, after quality controled, the files are updated and made available to researchers. There are currently a total of 17,324 files.

### Dimensions

Individual files contains two dimension:

- `['traj']`: 1, one trajectory per file (e.g. `ID`, `deploy_date`, `deploy_lon`, etc.)
- `['obs']`: X, with X the number of observations along the trajectory (`longitude`, `latitude`, `ve`, `vn`, `sst`, etc.)
 
### Variables

There are 20 variables measured (or postprocessed) along the trajectory, their dimension is `['traj', 'obs']` (`traj` = 1 per file). Naturally, metadata such as: the date of deployment, the type of buoy, etc., are constant along the trajectory and are dimension `['traj']`.

## Ragged Array

In the GDP dataset, the number of observations varies from `len(['obs'])=13` to `len(['obs'])=66417`. Because of this, it is **highly unefficient** to create a bidimensional datastructure `['traj', 'obs']`, commonly used in numerical Lagrangian Ocean Analysis toolbox such as [Ocean Parcels](https://oceanparcels.org/) and [OpenDrift](https://opendrift.github.io/). For example, the inclusion of hourly fitted sea surface temperature added 9 variables to the dataset (including diurnal, non diurnal, errors and interpolations flag). We can also expect new sensors and/or postprocessing methods will increase the size of the dataset in the future.

For this, reason we propose to combine the data from the individual netCDFs files into a [*contiguous ragged array*](https://cfconventions.org/cf-conventions/cf-conventions.html#_contiguous_ragged_array_representation), to simplify the distribution, decrease metadata redundancies, and efficiently store nonregular (in size) Lagrangian data. The preprocessing tools (`preprocess.py`) also converts to variables some of the Metadata initially stored as attributes in the individual netCDFs. The final structure contains 21 `vars['obs']` and 38 `vars['traj']`.

## *Xarray* test 1: Geographical binning of any variable
The first test is the bins statistic of meridional velocity of all drifters of the dataset. From the previous cell, we can see that the size of the chunks were automatically set to the size of `obs`.

For this type of operation, the size of the chunks is convenient because the *complete meridional velocity ragged array* will be loaded in memory in one operation. To perform operations per trajectory (shown below), the size of the chunks is set to the lengh of each trajectory, which would require `subset_nb_drifters` operation to load the same data. 

Rechunking is required as a function of the operations to perform.

```python
benchmark_times = np.zeros((3,3))  # 3 benchmark tests for the 3 different processing libraries
```

```python
t0 = time.time()

lon = np.linspace(-180, 180, 360*2)
lat = np.linspace(-90, 90, 180*2)

ret = stats.binned_statistic_2d(ds.longitude, 
                                ds.latitude,
                                ds.ve,
                                statistic=np.nanmean, 
                                bins=[lon, lat])

benchmark_times[0,0] = time.time() - t0
```

```python
x_c = np.convolve(lon, [0.5, 0.5], mode='valid')
y_c = np.convolve(lat, [0.5, 0.5], mode='valid')

fig = plt.figure(dpi=150)
ax = fig.add_subplot(1,1,1,projection=ccrs.Robinson(central_longitude=-180))
pcm = ax.pcolormesh(x_c, y_c, 
                    ret.statistic.T, 
                    cmap=cmocean.cm.balance, 
                    transform=ccrs.PlateCarree(),
                    vmin=-1.25, vmax=1.25)

# gridlines and labels
gl = ax.gridlines(color='k', linewidth=0.1, linestyle='-',
                  xlocs=np.arange(-180, 181, 60), ylocs=np.arange(-90, 91, 30),
                  draw_labels=True)
gl.top_labels = False
gl.right_labels = False

# add land and coastline
ax.add_feature(cfeature.LAND, facecolor='grey', zorder=1)
ax.add_feature(cfeature.COASTLINE, linewidth=0.25, zorder=1)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="3%", pad=0.05, axes_class=plt.Axes)
cb = fig.colorbar(pcm, cax=cax);
cb.ax.set_ylabel('Meridional velocity [m/s]');
```

## *Xarray* test 2: Extract a given region

Extracting a subregion of the dataset is also a very common operation. *Xarray* provides efficient way to apply mask to the data. One important thing to note is that since we have two dimensions, a mask is applied on the `obs` as well as on the `traj` dimensions.

```python
def retrieve_region(ds, lon: list = None, lat: list = None, time: list = None) -> xr.Dataset:
    '''Subset the dataset for a region in space and time
    
    Args:
        ds: xarray Dataset
        lon: longitude slice of the subregion
        lat: latitude slice of the subregion
        time: tiem slice of the subregion
    
    Returns: 
        ds_subset: Dataset of the subregion
    '''
    mask = np.ones(ds.dims['obs'], dtype='bool')

    if lon:
        mask &= (ds.coords['longitude'] >= lon[0]).values
        mask &= (ds.coords['longitude'] <= lon[1]).values

    if lat:
        mask &= (ds.coords['latitude'] >= lat[0]).values
        mask &= (ds.coords['latitude'] <= lat[1]).values

    if time:
        mask &= (ds.coords['time'] >= np.datetime64(time[0])).values
        mask &= (ds.coords['time'] <= np.datetime64(time[1])).values

    mask_id = np.in1d(ds.ID, np.unique(ds.ids[mask]))
    ds_subset = ds.isel(obs=np.where(mask)[0], traj=np.where(mask_id)[0])

    return ds_subset.compute()
```

```python
t0 = time.time()

# we need to record the time for benchmarking in the end
lon = [-98, -78]
lat = [18, 31]
day0 = '%sZ' % datetime(2000,1,1).isoformat()
day1 = '%sZ' % datetime(2020,12,31).isoformat()
days = [day0, day1]
ds_subset = retrieve_region(ds, lon, lat, days)

benchmark_times[0,1] = time.time() - t0
```

The the sea surface temperature in the extracted subregion, the Gulf of Mexico between 2020—2022, is plotted in the next cell.

```python
fig = plt.figure(dpi=150)
ax = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree())

pcm = ax.scatter(ds_subset.longitude, ds_subset.latitude, 
               s=0.05, c=ds_subset.sst-273.15, transform=ccrs.PlateCarree(),
               cmap=cmocean.cm.thermal)

ax.add_feature(cfeature.LAND, facecolor='grey', zorder=1)
ax.add_feature(cfeature.COASTLINE, linewidth=0.25, zorder=1)
ax.set_xticks(np.arange(-95, -79, 5), crs=ccrs.PlateCarree())
ax.set_yticks([20, 25, 30], crs=ccrs.PlateCarree())
ax.xaxis.set_major_formatter(LongitudeFormatter())
ax.yaxis.set_major_formatter(LatitudeFormatter())
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='3%', pad=0.02, axes_class=plt.Axes)
cb = fig.colorbar(pcm, cax=cax)
cb.set_label('Sea Surface Temperature [˚C]')
```

### Simple operation by chunks (or blocks)

For example, a function that calculates the `mean()`, `min()`, and `max()` of an array `x`. When mapped per trajectory (per blocks or chunks), the results will be combined as such: `[mean_traj_1, min_traj_1, max_traj_1, mean_traj_2, min_traj_2, max_traj_2, ..., mean_traj_n, min_traj_n, max_traj_n]` resulting in an `numpy array` 3 times as long.

The function is applied to each chunks/blocks here using `map_blocks()`. Note that this requires loading `nb_subset_drifters` chunks and applying the operation.

```python
%%time
def func(x):
    return np.array([np.nanmean(x)])

stats_traj = ds.sst.data.map_blocks(func).compute() # apply to sst variable
```

```python
stats_traj[:10]
```

### More complex operation where dimensions can change between input and output

Similarly, one can use `apply_ufunc` to map a function per chunks. The function is a bit more complex, but gives more flexibility to handle the input/output format of the function.

In this first example, the Eastward velocity is passed by chunk to the function which will calculate the anomaly per trajectory. 

- *Note 1*: the input and output sizes are not modified.
- *Note 2*: you must passed `ds.ve.data` (which retrieves the underlying dask array) to the function

```python
def per_chunk_anomaly(array):
    # MUST return an array (https://github.com/dask/dask/issues/8822)
    return array.map_blocks(lambda x: x-np.nanmean(x), chunks=array.chunks)
```

```python
%%time

ret = xr.apply_ufunc(
    per_chunk_anomaly,
    ds.ve.data,  # .data to retrieve the underlying chunked dask array
    input_core_dims=[['obs']],
    output_core_dims=[['obs']],
    dask='allowed'
).compute()
```

```python
ret
```

This second example calculates the `mean()` per trajectory. This means that for our datasets consisting of `nb_subset_drifters` chunks, the output will have `nb_subset_drifters` values or 1 value per chunk.

```python
def per_chunk_mean(array):
    nchunks = len(array.chunks[0])
    output_chunks = ([1] * nchunks,) # 1 value per chunk
    return array.map_blocks(lambda x: np.nanmean(x, keepdims=True), chunks=output_chunks)
```

```python
%%time

ret = xr.apply_ufunc(
    per_chunk_mean,
    ds.ve.data,
    input_core_dims=[['obs']],  # input dimension to the per_block function
    output_core_dims=[['obs']], # output still has one dimension
    exclude_dims=set(['obs']),  # size of x changes so it has to be in the exclude_dims param
    dask='allowed',
).compute()
```

```python
len(ret)
```

```python
ret[:10]
```

A last example, we calculate `fft` on `u+vi` which is a similar step involved in the calculation of *rotary spectrum*. The combinations of `apply_ufunc` and `map_blocks` is pretty powerful and can serve as templates for most processing applied per trajectory.

```python
ds = ds.assign(cv=ds['ve'] + 1j * ds['vn'])
ds.cv
```

```python
def per_chunk_fft(array):
    return array.map_blocks(lambda x: np.fft.fft(x), chunks=array.chunks)
```

```python
t0 = time.time()

ret = xr.apply_ufunc(
    per_chunk_fft,
    ds['cv'].data,
    input_core_dims=[["obs"]],  # input dimension to the per_block function
    output_core_dims=[["obs"]],  # output still has one dimension
    dask='allowed'
).compute()

benchmark_times[0,2] = time.time() - t0
```

## *Pandas* test 2: Extract a given region

The extraction of a region is slightly simpler than with *Xarray*, due to the fact that all variables have the `obs` dimension, so the mask can be directly applied.

```python
def retrieve_region_pd(df, lon: list = None, lat: list = None, time: list = None):
    '''Subset the dataset for a region in space and time
    
    Args:
        df: Pandas DataFrame
        lon: longitude slice of the subregion
        lat: latitude slice of the subregion
        time: tiem slice of the subregion
    
    Returns: 
        ds_subset: Dataset of the subregion
    '''
    mask = functools.reduce(np.logical_and, 
                            (
                                df.longitude > lon[0], 
                                df.longitude < lon[1],
                                df.latitude > lat[0],
                                df.latitude < lat[1],
                                df.time > day0,
                                df.time < day1,
                            )
                           )
    
    # this should work but crashes the kernel applying the where()
    return df.loc[mask]
```

# *Awkward Array*
[*Awkward Array*](https://awkward-array.readthedocs.io/en/latest/) is a library for nested, variable-sized data, including arbitrary-length lists, records, mixed types, and missing data, using NumPy-like idioms. Arrays are dynamically typed, but operations on them are compiled and fast. Their behavior coincides with NumPy when array dimensions are regular and generalizes when they’re not.

The main advantages of awkward array are:
- the indexing is already taken care of by the library, e.g. sst[400] returns the sea surface temperature along the 401$^{st}$ trajectory;
- ability to efficiently perform operations per trajectory;
- easy integration with numba.

```python
def retrieve_region_ak(ds: ak.Array, lon: list = None, lat: list = None, time: list = None) -> ak.Array:
    '''Subset the dataset for a region in space and time
    
    Args:
        ds: Awkward Array
        lon: longitude slice of the subregion
        lat: latitude slice of the subregion
        time: tiem slice of the subregion
    
    Returns: 
        ds_subset: Dataset of the subregion
    '''
    mask = functools.reduce(np.logical_and, 
                            (
                                ds.longitude > lon[0], 
                                ds.longitude < lon[1],
                                ds.latitude > lat[0],
                                ds.latitude < lat[1],
                                ds.time > day0,
                                ds.time < day1,
                            )
                           )
    
    keep = np.sum(mask == True, axis=1) > 0
    return ds[mask][keep]
```

## Numba

*Awkward Array* can be combined with Numba to accelerate calculations. Here, the goal is *not* to show the most efficient way of performing a calculation, but to present how simply adding the numba decorator `@nb.njit` can improved the performance by order of magnitude(s).

```python
def kinetic_energy(ke, u, v):
    for i in range(0, len(u)-1):
        ke[i] = np.sqrt(u[i]**2 + v[i]**2)
        
@nb.njit
def kinetic_energy_nb(ke, u, v):
    for i in range(0, len(u)-1):
        ke[i] = np.sqrt(u[i]**2 + v[i]**2)
```

```python
# select a random trajectory and preallocated the array
id_ex = 10
ke = np.zeros(len(ds_ak.obs.longitude[id_ex]))
ke2 = np.zeros_like(ke)
```

```python
%%time
kinetic_energy(ke, ds_ak.obs.longitude[id_ex], ds_ak.obs.latitude[id_ex])
```

The function is compiled just in time when we execute the function for the first time, future executions are *numba fast*.

```python
kinetic_energy_nb(ke2, ds_ak.obs.longitude[id_ex], ds_ak.obs.latitude[id_ex])
```

```python
%%timeit -r10 -n10
kinetic_energy_nb(ke2, ds_ak.obs.longitude[id_ex], ds_ak.obs.latitude[id_ex])
```

This results in more than a 1000x speedup (!) for the same function simply by adding the `@nb.njit` decorators!

# Discussion

We presented the GDP dataset and an efficient structure to combine variables from individual netCDF files into a ragged array. We then compare three different packages–xarray, Pandas, and Awkward Array—by performing typical Lagrangian workflow tasks.

**Xarray**

Xarray is a well known and established package used perform operation on N-Dimensional datasets. It has a great integration with Dask (not demonstrated in details in this Notebook), and allow for *lazy loading* when the dataset is larger than the available memory. 

However, it is not typically used with ragged array, so operation per trajectory requires to manually index through the dataset. One possibility, shown in the Xarray section, is to set the size of the Dask *chunks* equals to the size of the trajectories, which allows for mapping operation in parallel per chunk/block/trajectory. Although, the high number of chunks is less than ideal and hinder the performance since a simple `mean()` across a dataset containing *n* trajectories requires *n* reading operation.

**Pandas**

Pandas is a predominantly used python data analysis package. It is design to perform operations on a 2-dimensional list (row/columns). To represent the GDP datasets, this would requires two *DataFrame*: one containing the variables with one value per trajectory `vars['traj']`, and a second one containing the variables with one value per observations along the trajectory (`vars['obs']`). This is not a deal-breaker, but nonetheless creates a separation between the metadata and the data.

For this particular Notebook, which consist of a subset of the GDP dataset, the operations are performed *fast* and in a quite elegant and intuitive fashion. Although, our test on the complete dataset where much slower than xarray or Awkward Array which seems to manage memory much more efficiently.

```python
# print a summary of the system for log purpose
# inspired from https://stackoverflow.com/a/58420504/1558320
def getSystemInfo():
    info={}
    info['platform']=platform.system()
    info['platform-release']=platform.release()
    info['platform-version']=platform.version()
    info['architecture']=platform.machine()
    info['processor']=platform.processor()
    try:
        info['frequency']=psutil.cpu_freq(percpu=False)
    except:
        info['frequency']='N/A'
    info['cores']=psutil.cpu_count(logical=False)
    info['threads']=psutil.cpu_count(logical=True)
    info['ram']=str(round(psutil.virtual_memory().total / (1024.0 **3)))+" GB"
    return info

getSystemInfo()
```
