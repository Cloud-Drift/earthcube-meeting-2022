## Purpose

For data, “Lagrangian” refers to oceanic and atmosphere information acquired by observing platforms drifting with the flow they are embedded within, but also more broadly refers to the data originating from uncrewed platforms, vehicles, and animals that gather data along their unrestricted but complicated paths. Because such paths traverse both spatial and temporal dimensions, Lagrangian data often convolve spatial and temporal information that cannot always and readily be organized, cataloged, and stored in common data structures and file formats with the help of common libraries and standards. As such, for both data generators and data users, Lagrangian data present challenges that the [CloudDrift](https://github.com/Cloud-Drift) project aims to overcome.

This notebook consists of systematic comparisons and evaluations of workflows for Lagrangian data, using as a basis the velocity and sea surface temperature dataset emanating from the drifting buoys of the [Global Drifter Program](https://www.aoml.noaa.gov/phod/gdp/). Specifically, we consider the interplay between diverse storage file formats ([NetCDF](https://www.unidata.ucar.edu/software/netcdf/), [Parquet](https://github.com/apache/parquet-format)) and the data structure associated with common existing libraries in Python ([xarray](https://docs.xarray.dev/en/stable/), [pandas](https://pandas.pydata.org), and [awkward](https://awkward-array.org/quickstart.html)) in order to test their adequacies for performing three common Lagrangian tasks: (i) binning of a variable on an Eulerian grid (e.g. mean temperature map); (ii) extracting data within given geographical and/or temporal windows; and (iii) analyses per trajectory (e.g. single statistics, Fast Fourier Transforms). Because the CloudDrift project aims at accelerating the use of Lagrangian data for atmospheric, oceanic, and climate sciences, we hope that this notebook will incite the community of users to eventually test it on their own platform and thus provide us with feedback on its ease of use and the intuitiveness of the proposed methods and guide future developments.

# Data Overview

The [Global Drifter Program (GDP)](https://www.aoml.noaa.gov/phod/gdp/) produces two interpolated datasets of drifter position, velocity and sea surface temperature from more than 20,000 drifters that have been released since 1979. One dataset is at 6-hour resolution ([Hansen and Poulain 1996](http://dx.doi.org/10.1175/1520-0426(1996)013<0900:QCAIOW>2.0.CO;2)) and the other one is at hourly resolution ([Elipot et al. 2016](http://dx.doi.org/10.1002/2016JC011716)). In the GDP dataset, the number of observations varies from `len(['obs'])=13` to `len(['obs'])=66417`. ## Bench test 2: Extract a given region 

````python
def retrieve_region(ds, lon: list = None, lat: list = None, time: list = None):
    '''

    :param lon:
    :param lat:
    :param time:
    :return:
    '''
    mask = np.ones(ds.dims['obs'], dtype='bool')

    if lon:
        mask &= (ds.coords['longitude'] >= lon[0]).values
        mask &= (ds.coords['longitude'] <= lon[1]).values

    if lat:
        mask &= (ds.coords['latitude'] >= lat[0]).values
        mask &= (ds.coords['latitude'] <= lat[1]).values

    if time:
        mask &= (ds.coords['time'] >= np.datetime64(time[0])).values
        mask &= (ds.coords['time'] <= np.datetime64(time[1])).values

    mask_id = np.in1d(ds.ID, np.unique(ds.ids[mask]))
    ds_subset = ds.isel(obs=np.where(mask)[0], traj=np.where(mask_id)[0])

    return ds_subset.compute(), mask
````

## Applying operation per trajectory (per chunks) using `map_blocks` and the more versatile `apply_ufunc`

For example, a function that calculates the `mean()`, `min()`, and `max()` of an array `x`. When mapped per trajectory (per blocks or chunks), the results will be combined as such: `[mean_traj_1, min_traj_1, max_traj_1, mean_traj_2, min_traj_2, max_traj_2, ..., mean_traj_n, min_traj_n, max_traj_n]` resulting in an `numpy array` 3 times as long.

The function is applied to each chunks/blocks here using `map_blocks()`. Note that this requires loading 17324 chunks and applying the operation.

````python
%%time
# can we separate the cell between function definition and function application? and keep track of time
def func(x):
    return np.array([np.nanmean(x), np.nanmin(x), np.nanmax(x)])

stats_traj = ds.ve.data.map_blocks(func).compute() # apply the variable ve: eastward velocity

assert(len(stats_traj) == 3 * ds.dims['traj']) # three times the number of traj
````

## `apply_ufunc`

Similarly, one can use `apply_ufunc` to map a function per chunks. The function is a bit more complex, but gives more flexibility to handle the input/output format of the function.

In this first example, the Eastward velocity is passed by chunk to the function which will calculate the anomaly per trajectory. 

- *Note 1*: the input and output sizes are not modified.
- **Note 2**: you must passed `ds.ve.data` to the function

````python
%%time
# we need more explanation, and a better visualization of the results
def per_chunk_stats(array):
    # MUST return an array (https://github.com/dask/dask/issues/8822)
    return array.map_blocks(lambda x: x-np.nanmean(x), chunks=array.chunks)

xr.apply_ufunc(
    per_chunk_stats,
    ds.ve.data,  # .data to retrieve the underlying chunked dask array
    input_core_dims=[['obs']],
    output_core_dims=[['obs']],
    dask='allowed'
).compute()
````

This second example calculates the `mean()` per trajectory. This means that for our datasets consisting of 17324 chunks, the output will have 17324 values or 1 value per chunk.

````python
%%time

def per_chunk_mean(array):
    nchunks = len(array.chunks[0])
    output_chunks = ([1] * nchunks,) # 1 value per chunk
    return array.map_blocks(lambda x: np.nanmean(x, keepdims=True), chunks=output_chunks)

xr.apply_ufunc(
    per_chunk_mean,
    ds.ve.data,
    input_core_dims=[['obs']],  # input dimension to the per_block function
    output_core_dims=[['obs']], # output still has one dimension
    exclude_dims=set(['obs']),  # size of x changes so it has to be in the exclude_dims param
    dask='allowed',
).compute()
````

A last example, we calculate `fft` on `u+vi` which is a similar step involved in the calculation of *rotary spectrum*. The combinations of `apply_ufunc` and `map_blocks` is pretty powerful and can serve as templates for most processing applied per trajectory.

````python
#ds['cmp_uv'].data # ds has no 'cmp_uv' variable?? let's call that cv
ds = ds.assign(cv=ds['ve'] + 1j * ds['vn'])
ds.cv
````

````python
%%time
# save output of time? and output?
def per_chunk_fft(array):
    return array.map_blocks(lambda x: np.fft.fft(x), chunks=array.chunks)

xr.apply_ufunc(
    per_chunk_fft,
    ds['cv'].data,
    input_core_dims=[["obs"]],  # input dimension to the per_block function
    output_core_dims=[["obs"]],  # output still has one dimension
    dask='allowed'
).compute()
````

## Bench test 2: Extracting a given region

````python
def retrieve_region_ak(ds, lon: list = None, lat: list = None, time: list = None):
    
    mask = functools.reduce(np.logical_and, 
                            (
                                ds.longitude > lon[0], 
                                ds.longitude < lon[1],
                                ds.latitude > lat[0],
                                ds.latitude < lat[1],
                                ds.time > day0,
                                ds.time < day1,
                            )
                           )
    
    keep = np.sum(mask == True, axis=1) > 0
    return ds[mask][keep]
````

````python
%%time

lon = [-98, -78]
lat = [18, 31]
day0 = (datetime(2015,1,1) - datetime(1970,1,1)).total_seconds()
day1 = (datetime(2020,12,31) - datetime(1970,1,1)).total_seconds()
time = [day0, day1]
ds_subset = retrieve_region_ak(ds_ak.obs, lon, lat, time)
````

## Numba

Awkward array can be combined with Numba to accelerate calculations.

````python
@nb.jit
def fft_per_traj(uv):
    d = []
    for i in range(0, len(uv)):
        d.append(np.fft.fft(uv[i]))
    return d
````

````python
%%time

r = fft_per_traj(
    ds_ak.obs.ve + ds_ak.obs.vn*1j
)
````
